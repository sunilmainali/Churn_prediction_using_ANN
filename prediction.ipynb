```python
import tensorflow as tf
import pickle
from tensorflow.keras.models import load_model
import pandas as pd
import numpy as np
```

```python

model=load_model(r'C:\Users\sunil\LEARNING\Churn_prediction\churn_model.h5')

with open("le_gen.pkl","rb") as f1:
    le_gen=pickle.load(f1)
with open("ohe_geo.pkl","rb") as f2:
    ohe_geo=pickle.load(f2)

with open("scaler.pkl","rb") as f3:
    scaler=pickle.load(f3)
```

```python
ohe_geo
```

```python
model
```

```python
# Example input data
input_data = {
    'CreditScore': 600,
    'Geography': 'France',
    'Gender': 'Male',
    'Age': 40,
    'Tenure': 3,
    'Balance': 60000,
    'NumOfProducts': 2,
    'HasCrCard': 1,
    'IsActiveMember': 1,
    'EstimatedSalary': 50000
}
```

```python
# One-hot encode 'Geography'
geo_encoded = ohe_geo.transform([[input_data['Geography']]])
geo_encoded_df = pd.DataFrame(geo_encoded, columns=ohe_geo.get_feature_names_out(['Geography']))
geo_encoded_df
```

```python
input_df=pd.DataFrame([input_data])
input_df
```

```python
## Encode categorical variables
input_df['Gender']=le_gen.transform(input_df['Gender'])
input_df
```

```python
input_df=pd.concat([input_df.drop("Geography",axis=1),geo_encoded_df],axis=1)
input_df
```

```python
## Scaling the input data
input_scaled=scaler.transform(input_df)
input_scaled
```

```python
predict=model.predict(input_scaled)
predict
```

```python
if predict<0.5:
    print('The customer is likely to churn')
else:
    print('The customer is not likely to churn')
```
